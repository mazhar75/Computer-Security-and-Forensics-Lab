##RSA

```python
import time
import random
from Crypto.Util import number

def generate_prime(bits):
    return number.getPrime(bits)

def generate_keys(k):
    p = generate_prime(k//2)
    q = generate_prime(k//2)
    n = p * q
    phi = (p - 1) * (q - 1)

    e = 65537            # Commonly used public exponent
    d = pow(e, -1, phi)  # Modular multiplicative inverse of e mod phi

    return (e, n), (d, n)

def encrypt(message, public_key):
    e, n = public_key
    encrypted = [pow(ord(char), e, n) for char in message]
    return encrypted



def decrypt(encrypted, private_key):
    d, n = private_key
    decrypted = ''.join([chr(pow(char, d, n)) for char in encrypted])
    return decrypted


def main():
    # Step 1: Generate keys
    k = int(input("Enter key length (16, 32, 64, 96): "))
    start_time = time.time()
    public_key, private_key = generate_keys(k)
    key_generation_time = time.time() - start_time

    # Step 2: Input plain text
    plain_text = input("Enter plain text: ")

    # Step 3: Encrypt
    start_time = time.time()
    encrypted_text = encrypt(plain_text, public_key)
    encryption_time = time.time() - start_time

    # Step 4: Decrypt
    start_time = time.time()
    decrypted_text = decrypt(encrypted_text, private_key)
    decryption_time = time.time() - start_time

    # Step 5: Print results
    print("Bit Size =", k)
    print("Public Key: (e, n) =", public_key)
    print("Private Key: (d, n) =", private_key)
    print("Plain Text:\n", plain_text)
    print("Encrypted Text(ASCII):\n", encrypted_text)
    print("Decrypted Text:\n", decrypted_text)
    print("Execution Time:")
    print("Key Generation:", key_generation_time, "sec")
    print("Encryption Time:", encryption_time, "sec")
    print("Decryption Time:", decryption_time, "sec")

if __name__ == "__main__":
    main()
```
